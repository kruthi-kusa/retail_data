```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/kruth/Downloads/retail_data/data.csv", encoding='latin1')
df.head()
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna(subset='CustomerID')
```

```python
df.drop_duplicates()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
print(df['InvoiceDate'].dtype)
```

```python
df['Date'] = df['InvoiceDate'].dt.date
df['Time'] = df['InvoiceDate'].dt.time
```

```python
print(df[['InvoiceDate', 'Date', 'Time']].head())
```

```python
df.head()
```

```python
df['TotalPrice'] = df['Quantity']*df['UnitPrice']
```

```python
df=df.loc[df['Quantity']>0]
```

```python
df.head()
```

```python
#EDA
TotalRevenue = df['TotalPrice'].sum()
TotalRevenue
```

```python
#avg order val
avg_order_value = df.groupby('InvoiceNo')['TotalPrice'].sum().mean()
print(f"Average Order Value: ${avg_order_value:.2f}")
```

```python
#top 5 products by quantity sold
product_quantiy = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head()
product_quantiy
```
